```python
#############  子供という変数を作る
def male_female_child(passenger):
    age,sex = passenger
    if age < 16:
        return 'child'
    else:
        return sex

df['person'] = df[['Age','Sex']].apply(male_female_child,axis=1)
df.head(10)
```

```python
df['Sex'] = df['Sex'].apply(lambda x: 0 if x == 'male' else 1)
df.head(10)
## 性別をダミー化
```
